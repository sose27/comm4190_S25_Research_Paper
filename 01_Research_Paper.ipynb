## **II. Mood Management Theory: Background and Key Concepts**
Mood Management Theory (MMT) was developed by Dolf Zillmann in the 1980s to explain how people select media content as a way to regulate their emotional states. According to the theory, individuals are not passive consumers of media. Instead, they actively seek out content that helps them feel better, especially when experiencing negative emotions (Zillmann, 1988). At its core, MMT is based on the idea that people try to maximize pleasure and minimize discomfort, often without being fully aware of it.

Zillmann identified four main characteristics that influence how media affects mood regulation: hedonic valence, excitatory potential, absorption potential, and semantic affinity. 

1. **Hedonic valence** refers to the emotional tone of the content, such as whether it is uplifting, neutral, or distressing. Research shows that people in negative moods tend to choose content with positive hedonic valence to improve how they feel (Knobloch-Westerwick, 2006).
2. **Excitatory potential** describes how stimulating or calming the content is. For example, someone who is feeling anxious might look for something slow-paced and relaxing.
3. **Absorption potential** refers to how immersive the content is. Content with high absorption potential is more likely to pull the viewer in and distract them from their mood.
4. **Semantic affinity** reflects how closely the content relates to the person’s current emotional state. For instance, someone who is grieving might be drawn to a sad movie that mirrors their own experience.

One of the critiques of MMT is that it was developed in a media environment that looked very different from the one we have today. The theory was built around traditional forms of media like television and radio. It did not anticipate interactive technologies or on-demand, personalized content. In fact, its early foundations were influenced by behaviorist models such as operant conditioning, which focused on how animals learn to avoid discomfort. While this approach helped structure the theory, it also limits its relevance to newer communication technologies.

Even with those limitations, researchers have extended MMT into digital and social media contexts. Studies have looked at how people use streaming platforms, social media feeds, and online games as tools for emotional self-regulation (Knobloch-Westerwick & Alter, 2007). These environments give users much more control over their content, making mood-based choices more immediate and personalized. Now, with generative AI becoming more common, I believe MMT needs to be expanded again. Instead of just selecting existing content, users can prompt language models to create personalized responses or stories that match or improve their mood. This shift from choosing media to co-creating it introduces a new kind of mood management. It also positions LLMs not just as tools but as active participants in emotional communication.

## **III. Applying Mood Management Theory to LLMs**
If Mood Management Theory was originally concerned with how people select media to improve or maintain their emotional states, then LLMs challenge and expand that model by enabling users to generate content in direct response to their current mood. Rather than choosing from a fixed menu of television shows or songs, users can now prompt an LLM to produce something that is emotionally tailored, moment by moment. This changes the user from a passive selector to an active participant in their own mood management process.

LLMs allow users to express how they feel and receive content that aligns with or counteracts that state, depending on what they need. This makes them especially interesting to consider through the lens of MMT, which identifies four content-based mechanisms that influence emotional outcomes: hedonic valence, excitatory potential, absorption potential, and semantic affinity (Zillmann, 1988; Knobloch-Westerwick, 2006). All four of these can, in theory, be controlled through prompts and reflected in the responses that LLMs produce. This kind of emotionally responsive interaction is further supported by work from Reeves and Nass (1996), who found that people tend to treat computers and other media as if they were human conversational partners, especially when those systems use natural language and social cues.

For example, someone feeling overwhelmed might prompt ChatGPT for a calming story or a short meditation. The expected MMT-informed response would have low excitatory potential, high absorption potential, and positive hedonic valence. Someone feeling happy might request something energizing and humorous, with high excitatory potential and emotional congruence. What makes LLMs unique here is that they respond not only with text but also with tone, pacing, and emotional cues, all of which are drawn from the language used in the prompt and the model’s training data.

The idea of semantic affinity is also worth revisiting in this context. In traditional MMT applications, users often avoid media that is too closely related to their current mood if it is negative, since it can intensify the unwanted feeling. With LLMs, however, users sometimes request responses that mirror how they feel, such as asking for a sad poem while feeling sad. This suggests that mood management is more complex when interaction is involved, and LLMs offer a new space to test how semantic affinity operates in practice.

In sum, LLMs present a new kind of emotional interface. They enable people to shape the media they engage with in real time, which is different from simply selecting pre-existing content. If MMT is to remain relevant in the age of generative AI, it must adapt to account for this shift. The experiment I conducted takes a first step in exploring whether LLM-generated content aligns with the kinds of emotional responses that MMT would predict.
## **IV. Prompting Experiment: Testing Mood and LLM Responses**
To explore whether language models respond in ways consistent with Mood Management Theory, I conducted a small-scale experiment using structured prompts with Claude. I simulated three distinct emotional states (happiness, sadness, and stress) and observed how the model responded to each. My goal was to see whether Claude's responses aligned with MMT’s predictions: that people in negative moods will be drawn toward content with positive emotional tone, high absorption potential, and low excitatory intensity, while those in positive moods will prefer content that sustains their emotional state (Zillmann, 1988; Knobloch-Westerwick, 2006).

**I used the following three prompts:**
- "I’m feeling happy. Can you write me something that fits my mood?"


- "I’m feeling stressed. Can you suggest a show to help?"


- "I’m feeling sad. Can you suggest a song or something to help?"


Claude’s response to the happiness prompt was a gentle, poetic reflection on the nature of joy. It described how happiness sharpens sensory awareness and turns small moments into something meaningful. The language was warm and observant, and it invited the user to continue the interaction by requesting a poem or more specific writing. This response had positive hedonic valence, high absorption potential, and medium excitatory potential. It also had high semantic affinity, since it matched the happy mood rather than attempting to alter it.

When I prompted from a stressed emotional state, Claude responded with a list of calming TV shows, including Ted Lasso, The Great British Bake Off, and Our Planet. These suggestions aligned with what MMT would expect someone under stress to seek: content with positive hedonic tone, low excitatory potential, and high absorption (Knobloch-Westerwick & Alter, 2007). Claude also offered thoughtful comments about why each show might help, and concluded by asking about my genre preferences, showing that the interaction could continue in a personalized way.

For the sadness prompt, Claude provided song suggestions organized by emotional effect. Some songs were offered for comfort and uplift (Here Comes the Sun, Three Little Birds), others for cathartic emotional release (Fix You, Landslide), and some for mood improvement through energy (Good as Hell, Happy). It also included non-musical suggestions like taking a walk, drinking something warm, or writing down feelings. The range of content made it clear that the model was aiming to give the user a choice between mood matching and mood repair. From an MMT perspective, this was especially interesting, since it reflects the complexity of real emotional needs. In this case, hedonic valence ranged from neutral to positive, absorption potential was moderate, and semantic affinity was high.

## **V. Discussion**
The results of my prompting experiment suggest that LLMs like Claude are capable of generating content that aligns closely with Mood Management Theory. In each case, the model responded with emotionally appropriate material, offering suggestions or stories that either matched the user's mood or gently guided it toward a more positive state. This kind of responsiveness supports MMT’s idea that people seek out media to manage their affective state, but it also raises new questions about what it means to select content in an environment where that content is being generated in real time.

One of the most consistent patterns was Claude’s tendency to default to emotionally supportive and uplifting responses, even when the input mood was sad or stressed. For example, when asked to recommend songs for a sad mood, the model included not only comfort-oriented tracks like Here Comes the Sun but also options for emotional release like Fix You and energizing picks like Happy. This range reflects a more complex understanding of mood regulation, where people might seek either mood repair or mood matching depending on their emotional goals (Knobloch-Westerwick & Alter, 2007). The model also included non-musical suggestions, like taking a walk or writing down feelings, further emphasizing a holistic approach to mood management.

The same pattern appeared in the stressed condition, where Claude offered both passive, calming content like Weightless and more active, motivational tracks like Three Little Birds. In both cases, the model seemed to prioritize content that would elevate or stabilize the user’s mood, aligning with MMT’s prediction that people in negative states seek out positive, calming, and absorbing content (Zillmann, 1988). However, the inclusion of more intense, emotionally complex options suggests that LLMs can also support users who want to engage deeply with their current mood, rather than simply escape it.

The responses to the happy prompts were more straightforward, consistently focusing on content that would sustain a positive mood. This included high-energy songs like Dancing Queen and Walking on Sunshine, along with upbeat TV shows like Brooklyn Nine-Nine and Parks and Recreation. The model also generated personalized, reflective writing in response to the happiness prompt, demonstrating its capacity to not just mirror a user’s mood but also deepen it through emotionally resonant language. This kind of tailored, mood-consistent output reflects a more conversational, interactive form of mood management, where the user is not just selecting pre-existing content but actively co-creating their media experience.

Overall, these findings suggest that LLMs have the potential to serve as effective mood management tools, offering both immediate emotional relief and deeper, more reflective engagement depending on the user’s needs. However, the model’s strong tendency to default to positivity, even in response to sad or stressed inputs, raises questions about how well LLMs can support more complex or negative emotional states over time. This balance between emotional support and authentic mood reflection is an area where future research could provide valuable insights.

## **VI. Conclusion**
This paper explored how Mood Management Theory can help explain the ways people interact with LLMs like Claude when they are trying to regulate their emotions.
